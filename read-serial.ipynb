```python
import serial
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq, rfft
import numpy as np
%matplotlib inline
```

```python
ser = serial.Serial('/dev/ttyACM0', 115200)
# ser = serial.Serial('/dev/ttyUSB0', 9600)

y = []
x = []
N = 3840
count = 0

while count < N:
    cc=str(ser.readline())
    newCC = cc[2:][:-5]
    # print(newCC)
    y.append(float(newCC))
    x.append(count)
    count = count + 1
```

```python
# plt.stem(x, y, use_line_collection=True)
# print(y[200:])
# print(x[10:30])
plt.rcParams["figure.figsize"] = (32,3)
plt.plot(x, y,  '-')
plt.show()
```

```python
T = 1/8333
yf = fft(x)
xf = fftfreq(N, T)
mascara = xf > 0
xfAbs = 2.0 * np.abs(yf/N)
plt.plot(xf[mascara], xfAbs[mascara])
plt.ylabel("Amplitude")
plt.xlabel("Frequncia (Hz)")
plt.show()
```
