```python
import serial
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq, rfft
import numpy as np
%matplotlib inline
```

```python
def graph_spectrogram(data, name):
    fig,ax = plt.subplots(1)
    fig.subplots_adjust(left=0,right=1,bottom=0,top=1)
    ax.axis('off')
    pxx, freqs, bins, im = ax.specgram(x=data, Fs=8333)
    ax.axis('off') 
    fig.savefig('./dataset/training/capela_obstruida/' + name)
```

```python
ser = serial.Serial('/dev/ttyACM0', 115200)
# ser = serial.Serial('/dev/ttyUSB0', 9600)

y = []
x = []
N = 3840
I = 500

count = 0
countSet = 0

while countSet < I:
    while count < N:
        # print('amostrando.... ' + str(N))
        cc=str(ser.readline())
        newCC = cc[2:][:-5]
        # print(newCC)
        y.append(float(newCC))
        x.append(count)
        count += 1

    graph_spectrogram(y, 'capela_obstruida.' + str(countSet) + '.png')
    countSet += 1
    y = []
    x = []
    count = 0
```

```python
plt.rcParams["figure.figsize"] = (10,3)
plt.plot(x[100:500], y[100:500],  '-')
plt.show()
```
